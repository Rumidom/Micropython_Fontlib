```python
from PIL import Image
from IPython.display import display
```

```python
def UpscaleFont(fontpath,factor=2): 
    im = Image.open(fontpath)
    width, height = im.size
    outwidth = width*factor
    outheight = height*factor
    out_im = im.resize((outwidth,outheight), resample=Image.BOX)
    return out_im

def Convert_to_1bit(fontpath):
    img = Image.open(fontpath)
    img_1bit = img.convert('1')
    img_1bit.save(fontpath, bit=1)
```

```python
upscaled = UpscaleFont('fonts/IBM BIOS (8,8).bmp') 
display(upscaled)
```

```python
upscaled.save("output.bmp")
```
